```python
# Import Library

import pandas as pd
import numpy as np
from scipy.stats.mstats import winsorize
```

```python
# Upload Dataset

df = pd.read_csv("https://drive.google.com/uc?id=1LfQWProB0VjWN5q8bKuRIgn-stULfIRo")
df.info()
df.describe()
```

```python
# Cek missing value awal

print("Missing Value Awal:")
print(df.isnull().sum())
```

```python
# Cek missing value awal

print("Missing Value Awal:")
print(df.isnull().sum())
```

```python
# Menghapus Data Duplikat

df.drop_duplicates(inplace=True)
print("\nShape setelah hapus duplikat:", df.shape)
```

```python
# Normalisasi String

df['kota'] = df['kota'].str.strip().str.title()
df['kondisi'] = df['kondisi'].str.strip().str.lower()
```

```python
# Imputasi Missing Value

df['luas_m2'] = df['luas_m2'].fillna(df['luas_m2'].median())
df['harga_juta'] = df['harga_juta'].fillna(df['harga_juta'].median())
df['kamar'] = df['kamar'].fillna(df['kamar'].mode()[0])
```

```python
# Menangani Outlier (IQR)

for col in ['harga_juta', 'luas_m2', 'tahun_bangun']:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    batas_bawah = Q1 - (1.5 * IQR)
    batas_atas = Q3 + (1.5 * IQR)

    df[col] = df[col].clip(batas_bawah, batas_atas)
```

```python
# Validasi Data

assert df.isnull().sum().sum() == 0, "Masih ada missing value!"
assert df.duplicated().sum() == 0, "Masih ada data duplikat!"

print("\nShape Akhir:", df.shape)

# Simpan dataset bersih
df.to_csv("housing_clean.csv", index=False)

print("Dataset bersih berhasil disimpan!")
```
