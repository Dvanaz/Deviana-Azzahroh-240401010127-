```python
# IMPORT LIBRARY

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# LOAD DATASET

df = sns.load_dataset('titanic')

print("Shape Dataset :", df.shape)

display(df.head())
```

```python
# PEMILIHAN FITUR

cols = [
    'pclass',
    'sex',
    'age',
    'sibsp',
    'parch',
    'fare',
    'embarked',
    'survived'
]

df = df[cols].copy()

print("Shape Setelah Seleksi Kolom:")
print(df.shape)
```

```python
# CEK MISSING VALUE

print("Jumlah Missing Value:\n")
print(df.isnull().sum())
```

```python
# DISTRIBUSI TARGET

print("Distribusi Survived:\n")

print(
    df['survived']
    .value_counts(normalize=True)
    .round(3)
)
```

```python
# MENANGANI MISSING VALUE

# Age → Median
df['age'] = df['age'].fillna(
    df['age'].median()
)

# Embarked → Modus
df['embarked'] = df['embarked'].fillna(
    df['embarked'].mode()[0]
)

print("Missing Value Setelah Cleaning:\n")
print(df.isnull().sum())
```

```python
# ONE HOT ENCODING

df = pd.get_dummies(
    df,
    columns=['sex', 'embarked'],
    drop_first=True,
    dtype=int
)

print("Kolom Setelah Encoding:\n")
print(df.columns.tolist())
```

```python
# FEATURE DAN TARGET

X = df.drop('survived', axis=1)

y = df['survived']

print("Shape Feature :", X.shape)
print("Shape Target  :", y.shape)
```

```python
# TRAIN TEST SPLIT

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print(f"Jumlah Data Train : {X_train.shape[0]}")
print(f"Jumlah Data Test  : {X_test.shape[0]}")
```

```python
# PROPORSI TARGET TRAIN DAN TEST

print("Proporsi Survived pada Train:\n")

print(
    y_train.value_counts(normalize=True)
    .round(3)
)

print("\nProporsi Survived pada Test:\n")

print(
    y_test.value_counts(normalize=True)
    .round(3)
)
```

```python
# STANDARD SCALER

from sklearn.preprocessing import StandardScaler

num_cols = [
    'pclass',
    'age',
    'sibsp',
    'parch',
    'fare'
]

scaler = StandardScaler()

X_train[num_cols] = scaler.fit_transform(
    X_train[num_cols]
)

X_test[num_cols] = scaler.transform(
    X_test[num_cols]
)

print("Mean Training:")
print(scaler.mean_.round(2))

print("\nStandar Deviasi:")
print(scaler.scale_.round(2))
```
