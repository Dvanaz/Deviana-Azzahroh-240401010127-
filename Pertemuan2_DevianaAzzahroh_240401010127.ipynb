```python
# Q1 - Informasi Dataset Titanic

import pandas as pd

url_data = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic = pd.read_csv(url_data)

jumlah_baris, jumlah_kolom = titanic.shape

print(f"Jumlah baris : {jumlah_baris}")
print(f"Jumlah kolom : {jumlah_kolom}")

print("\nDaftar Kolom Dataset Titanic:")
for no, kolom in enumerate(titanic.columns, start=1):
    print(f"{no}. {kolom}")
```

```python
# Q2 - Missing Value

missing = titanic.isnull().sum()
persentase = (missing / len(titanic) * 100).round(2)

data_missing = pd.DataFrame({
    "Jumlah Missing": missing,
    "Persentase (%)": persentase
})

data_missing = data_missing[data_missing["Jumlah Missing"] > 0]

print("Kolom yang memiliki Missing Value:\n")

for i, (kolom, row) in enumerate(data_missing.iterrows(), start=1):
    print(f"{i}. {kolom}")
    print(f"   Jumlah Missing : {row['Jumlah Missing']}")
    print(f"   Persentase     : {row['Persentase (%)']}%")
    print()
```

```python
# Q3 - Persentase Penumpang Selamat

selamat = titanic['Survived'].value_counts()[1]
total = len(titanic)

persen_selamat = round((selamat / total) * 100, 2)

print(f"Jumlah selamat : {selamat}")
print(f"Persentase selamat : {persen_selamat}%")
```

```python
# Q4 - Wanita Kelas 1

female_first_class = titanic[
    (titanic['Sex'] == 'female') &
    (titanic['Pclass'] == 1)
]

jumlah_wanita = female_first_class.shape[0]
usia_rata = round(
    female_first_class['Age'].mean(), 2
)

print("Jumlah Penumpang :", jumlah_wanita)
print("Rata-rata Usia :", usia_rata)
```

```python
# Q5 - Tingkat Keselamatan per Kelas

keselamatan_kelas = (
    titanic.groupby('Pclass')['Survived']
    .mean()
    .mul(100)
    .round(2)
)

print("Persentase Keselamatan Berdasarkan Kelas:\n")

for kelas, persen in keselamatan_kelas.items():
    print(f"Kelas {kelas} : {persen}%")

print(f"\nKelas dengan tingkat keselamatan tertinggi adalah Kelas {keselamatan_kelas.idxmax()} dengan persentase {keselamatan_kelas.max()}%")
```

```python
# Q6 - Analisis Berdasarkan Embarked

embarked_analysis = (
    titanic.groupby('Embarked')['Survived']
    .mean()
    .mul(100)
    .round(2)
)

print(embarked_analysis)
```
